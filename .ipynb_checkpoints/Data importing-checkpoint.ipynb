## Drop Database if Exists

```python
##################### Use when you want to drop a database ###################
import mysql.connector as msql
from mysql.connector import Error
try:
    conn = msql.connect(host='localhost', user='root',  
                    password=getpass.getpass('Enter password:'))
    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("DROP DATABASE IF EXISTS Airbnb")
        print("Database is dropped")
except Error as e:
    print("Error while connecting to MySQL", e)
```

## Create Database

```python
import mysql.connector as msql
from mysql.connector import Error
try:
    conn = msql.connect(host='localhost', user='root',  
                    password=getpass.getpass('Enter password:'))
    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("CREATE DATABASE Airbnb_Seattle")
        print("Database is created")
except Error as e:
    print("Error while connecting to MySQL", e)
```

## Importing Calendar csv to MySQL Databasee

```python
try:
    conn = msql.connect(host='localhost', 
                        database='Airbnb',
                        user='root',  
                        password=getpass.getpass('Enter password:'))
                        
    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("select database();")
        
        
        
        record = cursor.fetchone()
        print("You're connected to database: ", record)
        cursor.execute('DROP TABLE IF EXISTS calendar;')
        print('Creating table....')
                        
#       pass the create table statement 
        cursor.execute("CREATE TABLE calendar(listing_id int NOT NULL, \
        date date DEFAULT NULL,available TEXT DEFAULT NULL, price text default NULL)")
    
        print("Table is created....")
        
         #loop through the data frame
        for i,row in calendar.iterrows():
            #here %S means string values 
            sql = "INSERT INTO Airbnb.calendar VALUES (%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))
            print("Record inserted")
            # the connection is not auto committed by default, so we must commit to save our changes
            conn.commit()
except Error as e:
            print("Error while connecting to MySQL", e)
```

## Importing Reviews csv to MySQL Database

```python
try:
    conn = msql.connect(host='localhost', 
                        database='Airbnb_Seattle', 
                        user='root',  
                        password=getpass.getpass('Enter password:'))
    if conn.is_connected():
        cursor = conn.cursor()
        cursor.execute("select database();")
        
        
        
        record = cursor.fetchone()
        print("You're connected to database: ", record)
        cursor.execute('DROP TABLE IF EXISTS reviews;')
        print('Creating table....')

#       pass the create table statement
        cursor.execute("CREATE TABLE reviews(`listing_id` int NOT NULL,\
        `id` int,\
        `date` date,\
        `reviewer_id` int DEFAULT NULL,\
        `reviewer_name` varchar(100) DEFAULT NULL)")
    
        print("Table is created....")
        
         #loop through the data frame
        for i,row in reviews.iterrows():
            
            # insert values
            sql = "INSERT INTO Airbnb_Seattle.reviews VALUES (%s,%s,%s,%s,%s)"
            cursor.execute(sql, tuple(row))
            print("Record inserted")
            
            conn.commit()
except Error as e:
            print("Error while connecting to MySQL", e)
```
